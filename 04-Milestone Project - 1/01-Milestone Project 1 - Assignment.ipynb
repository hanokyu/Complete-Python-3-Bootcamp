```python
# PROJECT: TTT
import random

ind = {'n1': ' ', 'n2': ' ', 'n3': ' ', 'n4': ' ', 'n5': ' ', 'n6': ' ', 'n7': ' ', 'n8': ' ', 'n9': ' '}


def print_board():
    print(' {n7} | {n8} | {n9} '.format(**ind))
    print('___|___|___')
    print(' {n4} | {n5} | {n6} '.format(**ind))
    print('___|___|___')
    print(' {n1} | {n2} | {n3} '.format(**ind))
    print('   |   |   ')


def win():
    return (ind['n1'] == ind['n2'] == ind['n3'] != ' ') or (
            ind['n4'] == ind['n5'] == ind['n6'] != ' ') or (
                   ind['n7'] == ind['n8'] == ind['n9'] != ' ') or (
                   ind['n1'] == ind['n4'] == ind['n7'] != ' ') or (
                   ind['n2'] == ind['n5'] == ind['n8'] != ' ') or (
                   ind['n3'] == ind['n6'] == ind['n9'] != ' ') or (
                   ind['n1'] == ind['n5'] == ind['n9'] != ' ') or (
                   ind['n3'] == ind['n5'] == ind['n7'] != ' ')


def draw():
    return not ' ' in ind.values() and not win()


def clearb():
    global ind
    ind = {'n1': ' ', 'n2': ' ', 'n3': ' ', 'n4': ' ', 'n5': ' ', 'n6': ' ', 'n7': ' ', 'n8': ' ', 'n9': ' '}


def chang(vt, xo):
    '''
    Tu input cua nguoi choi dat X/O tuong ung vao bang, in bang.
    '''
    vtri = {1: 'n1', 2: 'n2', 3: 'n3', 4: 'n4', 5: 'n5', 6: 'n6', 7: 'n7', 8: 'n8', 9: 'n9'}
    ind[vtri[vt]] = xo
    print('\n' * 100)
    print_board()


print('Chao mung den voi TIC TAC TOE!!\n')
bd = input('Nhap Y de bat dau choi!\n')
while bd.lower() != 'y':
    bd = input('\nChoi di ma :(\n')

# Lay ten va X/O

name1 = input('\nTen nguoi choi 1 la gi?\n')
name2 = input('\nTen nguoi choi 2 la gi?\n')

p1 = input(f'\nNoice, {name1} chon X hay O?\n')
while p1.lower() != 'x' and p1.lower() != 'o':
    p1 = input('\nNhap X hoac O thoi!\n')

if p1.lower() == 'x':
    p1 = 'X'
    p2 = 'O'
else:
    p2 = 'X'
    p1 = 'O'

input(f'\n{name1}: {p1}\n{name2}: {p2}\n')


def rule():
    print('Nhap so tuong ung voi vi tri trong bang de dat X/O.\nNguoi dau tien co 3 X/O thanh hang ngang/doc/cheo se thang!\n')
    global ind
    ind = {'n1': 1, 'n2': 2, 'n3': 3, 'n4': 4, 'n5': 5, 'n6': 6, 'n7': 7, 'n8': 8, 'n9': 9}
    print_board()


rule()
input('\nHieu luat choi chua?    ')
input('\nPhai hieu! Nhan enter de bat dau.')
print('\n'*100)

def go_f():
    if random.randint(1,2) == 1:
        return p1
    else:
        return p2

def play():
    move = 1
    move1 = []
    move2 = []
    clearb()
    print_board()
    
    first = go_f()
    second = ''
    if first == p1:
        print(f'\n{name1} di truoc ({p1})')
        second = p2
    else:
        print(f'\n{name2} di truoc ({p2})')
        second = p1
    
    while not win() and not draw():

        if move % 2 != 0 and not win():  # di truoc
            while True:
                try:
                    if p1 == first:
                        x1 = int(input(f'Luot cua {name1}:  '))  # Nhap dung so 1 - 9 (va ko trung)
                    else:
                        x1 = int(input(f'Luot cua {name2}:  '))
                    while x1 in move1 or x1 in move2 or not x1 in range(1, 10):
                        x1 = int(input(f'Khong hop le! Chon vi tri khac:  '))
                except ValueError:
                    print('Nhap so 1 - 9 co ma!')
                else:
                    break

            move1.append(x1)
            chang(move1[-1], first)
            move += 1

        if move % 2 == 0 and not win() and move < 10:  # di sau
            while True:
                try:
                    if p1 == second:
                        x2 = int(input(f'Luot cua {name1}:  '))  # Nhap dung so 1 - 9 (va ko trung)
                    else:
                        x2 = int(input(f'Luot cua {name2}:  '))
                    while x2 in move1 or x2 in move2 or not x2 in range(1, 10):
                        x2 = int(input(f'Khong hop le! Chon vi tri khac:  '))
                except ValueError:
                    print('Nhap so 1 - 9 co ma!')
                else:
                    break

            move2.append(x2)
            chang(move2[-1], second)
            move += 1

    if win():
        if (move - 1) % 2 != 0:  # 1 thang
            if first == p1:
                print(f'Chuc mung! {name1} da thang!')
            else:
                print(f'Chuc mung! {name2} da thang!')
        else:  # 2 thang
            if second == p2:
                print(f'Chuc mung! {name2} da thang!')
            else:
                print(f'Chuc mung! {name1} da thang!')

    else:
        print('Hoa!!')

play()


def reset():
    '''
    Don bang, lay ten, hien luat
    '''
    clearb()
    global name1, name2, p1, p2
    name1 = input('\nTen nguoi choi 1 la gi?\n')
    name2 = input('\nTen nguoi choi 2 la gi?\n')

    p1 = input(f'\nNoice, {name1} chon X hay O?\n')
    while p1.lower() != 'x' and p1.lower() != 'o':
        p1 = input('\nNhap X hoac O thoi!\n')

    if p1.lower() == 'x':
        p1 = 'X'
        p2 = 'O'
    else:
        p2 = 'X'
        p1 = 'O'

    input(f'\n{name1}: {p1}\n{name2}: {p2}\n')
    rule()
    input('Nhan enter:  ')
    print('\n' * 100)


def replay():
    '''
    Hoi nguoi choi co choi tiep khong
    '''
    rp = input('Choi tiep khong? (Y/N)\n')
    while rp.lower() != 'y' and rp.lower() != 'n':
        rp = input('The la co hay khong -_-\n')
    return rp.lower() == 'y'


while True:
    if not replay():
        input('\nBye-bye!')
        break
    else:
        print('\nNoice!\n')
        reset()
        play()

# End of program :)
```
